1. Standardize column names to make them consistent (e.g., lowercase, no spaces)

```python
import pandas as pd

# Correct the file path using one of the methods above
file_path = r"C:\Users\pavan\Desktop\Medini projects\medical_data\medical_data_analysis\dataset\healthcare_dataset.csv"

# Load the CSV file
data = pd.read_csv(file_path)

# Standardize column names
data.columns = [col.lower().replace(" ", "_") for col in data.columns]

# Save the updated CSV file
output_file = "standardized_healthcare_dataset.csv"
data.to_csv(output_file, index=False)

print(f"Standardized file saved as: {output_file}")
```

```python
import pandas as pd

# Load the CSV file
file_path = r"C:\Users\pavan\Desktop\Medini projects\medical_data\medical_data_analysis\dataset\healthcare_dataset.csv"
  # Replace with your CSV file path
data = pd.read_csv(file_path)

# Specify the desired data types for specific columns (update as per your dataset)
# Example: {'date_column': 'datetime', 'numeric_column': 'float', 'integer_column': 'int'}
columns_to_convert = {
    "Billing Amount": "float",   # Replace 'numeric_column' with actual column name
    "Room Number": "int"      # Replace 'integer_column' with actual column name
}

# Validate and convert data types
for col, dtype in columns_to_convert.items():
    if col in data.columns:
        if dtype == "datetime":
            data[col] = pd.to_datetime(data[col], errors="coerce")  # Coerce invalid dates to NaT
        elif dtype in ["float", "int"]:
            data[col] = pd.to_numeric(data[col], errors="coerce").astype(dtype, errors="ignore")  # Convert to numeric

# Handle missing data if needed (e.g., fill NaN or drop rows)
# Example: data.fillna(0, inplace=True) or data.dropna(inplace=True)

# Save the updated CSV file
output_file = "standardized_healthcare_dataset.csv"
data.to_csv(output_file, index=False)

print(f"Validated and updated file saved as: {output_file}")
```

2.Analyze demographic distribution (age, gender).

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r"C:\Users\pavan\Desktop\Medini projects\medical_data\medical_data_analysis\dataset\healthcare_dataset.csv"  # Replace with your CSV file path
data = pd.read_csv(file_path)

# Analyze Age Distribution
if "Age" in data.columns:
    print("\nAge Distribution Statistics:")
    print(data["Age"].describe())
    data["Age"].plot(kind="hist", bins=20, title="Age Distribution", color="skyblue", edgecolor="black")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()
else:
    print("\nAge column not found in the dataset.")

# Analyze Gender Distribution
if "Gender" in data.columns:
    print("\nGender Distribution:")
    gender_counts = data["Gender"].value_counts()
    print(gender_counts)
    gender_counts.plot(kind="bar", title="Gender Distribution", color=["lightblue", "pink"])
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()
else:
    print("\nGender column not found in the dataset.")
```

3 Examine the distribution of medical conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r"C:\Users\pavan\Desktop\Medini projects\medical_data\medical_data_analysis\dataset\healthcare_dataset.csv"  # Replace with your CSV file path
data = pd.read_csv(file_path)

# Analyze Medical Conditions Distribution
if "Medical Condition" in data.columns:
    print("\nDistribution of Medical Condition:")
    condition_counts = data["Medical Condition"].value_counts()
    print(condition_counts)

    # Plot the distribution of medical conditions
    condition_counts.plot(kind="barh", title="Distribution of Medical Condition", color="skyblue", edgecolor="black")
    plt.xlabel("Count")
    plt.ylabel("Medical Condition")
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.show()
else:
    print("\nThe column 'medical_conditions' was not found in the dataset.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r"C:\Users\pavan\Desktop\Medini projects\medical_data\medical_data_analysis\dataset\healthcare_dataset.csv"  # Replace with your CSV file path
data = pd.read_csv(file_path)

# Analyze Hospital-Wise Patient Distribution
if "Hospital" in data.columns:  # Replace with the actual column name if different
    print("\nHospital-Wise Patient Distribution:")
    hospital_counts = data["Hospital"].value_counts()
    print(hospital_counts)

    # Plot the distribution of patients across hospitals
    hospital_counts.plot(kind="bar", title="Hospital-Wise Patient Distribution", color="skyblue", edgecolor="black")
    plt.xlabel("Hospital Name")
    plt.ylabel("Number of Patients")
    plt.xticks(rotation=45, ha="right")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.tight_layout()  # Adjust layout to prevent label cutoff
    plt.show()
else:
    print("\nThe column 'hospital_name' was not found in the dataset.")
```
